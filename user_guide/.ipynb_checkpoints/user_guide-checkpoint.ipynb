<center><b><font size='6'><u>Hierarchical Bayesian Inference Tool for Knowledge DAGs</u></font></b> </center>

<b>Hierarchical Bayesian Inference Tool for Knowledge DAGs</b> is a tool for inferring individuals' causal scientific knowledge from the individuals' responses to a set of questions.
This tool is based on a probabilistic graph-based logistic (GrL) model that uses directed acyclic graphs (DAGs) to represent individuals' causal scientific knowledge for a given theory.
The code has been written in Python. 
Below are the steps to run the tool efficiently.

+ First you need to start the "knowledge_DAGs.ipynb" jupyter notebook file which can be found in the "main" folder.
+ Once "knowledge_DAGs.ipynb" is open, click on Cell->Run All. 
+ The output of the last cell should be:

+ Steps below will guide you on the specifics of data required to run this tool

<font size='4'><b>1) Loading Data</b></font>
    
1. This section is marked in the tool as: <b>1) Theory Specific Inputs</b> and is shown in the figure below: 


2. There are 5 input files required to run this tool: Nodes, Edges, Sub-Graphs Nodes, Question-Specific Nodes and Question Responses. A sample of each of these input file can be found in the Sample_Data directory. It is <b> EXTREMELY!!!</b> important to follow the formatting similar to that of the files in the Sample_Data directory.

    2.1 The file corresponding to the input widget, Nodes, correspond to the vertices of the directed acyclic graph (DAG). The file format <b>MUST</b> be (.txt).
    
    2.2 The file corresponding to the input widget, Edges, correspond to the edges between the vertices of the directed acyclic graph (DAG). For example, in the sample file: Edges.txt in the Sample_Data directory, an edge ('G','M') corresponds to an edge from vertices 'G' to 'M'. The file format <b>MUST</b> be (.txt). 
    
    2.3 The file corresponding to the input widget, Sub-Graphs Nodes, correspond to the vertices of the directed acyclic graph (DAG) that lie in the same sub-graph. For example, in the sample file: Sub-Graph-Nodes.txt in the Sample_Data directory, vertices ['ka','kb','kc','kd','ke'] lie in the same sub-graph. This assumes that an individual knows all the vertices in this sub-graph with the same probability. <b>IF YOU DO NOT WANT TO DEFINE SEPERATE SUB-GRAPHS AND HAVE (or WANT) DIFFERENT PROBABILITY FOR EVERY LINK BETWEEN NODES, JUST UPLOAD THE SAME 'NODES' FILE AS IN (2.1)</b>. The file format <b>MUST</b> be (.txt). 
    
    2.4 The file corresponding to the input widget, Question-Specific Nodes, correspond to the set of vertices that are required to answer a given question. For example, the sample file: Question-Specific-Nodes.txt in the Sample_Data directory, contains 10 set of vertices corresponding to 10 different question. In this sample file, the first set, ['G', 'F','M','sig_o'] correspond to the vertices relevant to answer question 1. The file format <b>MUST</b> be (.txt). 
    
    2.5 The file corresponding to the input widget, Question Responses, contains individuals' responses to the questions uploaded in (2.4). For example, in the sample file: Question-Responses.csv in the Sample_Data directory, there are 6 rows and 10 columns. This corresponds to response of 6 individuals to 10 questions. The file format <b>MUST</b> be (.csv).
